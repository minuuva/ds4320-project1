```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import logging
import sys

# Setup comprehensive logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('pipeline_demo.log', mode='w'),
        logging.StreamHandler(sys.stdout)
    ]
)

logging.info("="*70)
logging.info("MOVIELENS 25M PIPELINE DEMONSTRATION STARTED")
logging.info("="*70)

# Data directory
data_dir = '../../data_verification/ml-25m/'
logging.info(f"Data directory: {data_dir}")
```

```python
# Initialize storage for tables
tables = {}

try:
    # Load movies
    logging.info("\n[1/7] Loading movies.csv...")
    tables['movies'] = pd.read_csv(f'{data_dir}movies.csv')
    logging.info(f"  ✓ Loaded {len(tables['movies']):,} movies")
    
except FileNotFoundError as e:
    logging.error(f"  ✗ Movies file not found: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ Error loading movies: {e}")
    raise

try:
    # Load ratings
    logging.info("\n[2/7] Loading ratings.csv...")
    tables['ratings'] = pd.read_csv(f'{data_dir}ratings.csv')
    logging.info(f"  ✓ Loaded {len(tables['ratings']):,} ratings")
    
except FileNotFoundError as e:
    logging.error(f"  ✗ Ratings file not found: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ Error loading ratings: {e}")
    raise

try:
    # Load users
    logging.info("\n[3/7] Loading users.csv...")
    tables['users'] = pd.read_csv(f'{data_dir}users.csv')
    logging.info(f"  ✓ Loaded {len(tables['users']):,} users")
    
except FileNotFoundError as e:
    logging.error(f"  ✗ Users file not found: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ Error loading users: {e}")
    raise

try:
    # Load tags
    logging.info("\n[4/7] Loading tags.csv...")
    tables['tags'] = pd.read_csv(f'{data_dir}tags.csv')
    logging.info(f"  ✓ Loaded {len(tables['tags']):,} tags")
    
except FileNotFoundError as e:
    logging.error(f"  ✗ Tags file not found: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ Error loading tags: {e}")
    raise

try:
    # Load links
    logging.info("\n[5/7] Loading links.csv...")
    tables['links'] = pd.read_csv(f'{data_dir}links.csv')
    logging.info(f"  ✓ Loaded {len(tables['links']):,} links")
    
except FileNotFoundError as e:
    logging.error(f"  ✗ Links file not found: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ Error loading links: {e}")
    raise

try:
    # Load genome scores (sample for performance)
    logging.info("\n[6/7] Loading genome-scores.csv (sample)...")
    tables['genome_scores'] = pd.read_csv(f'{data_dir}genome-scores.csv', nrows=100000)
    logging.info(f"  ✓ Sampled {len(tables['genome_scores']):,} genome scores")
    
except FileNotFoundError as e:
    logging.error(f"  ✗ Genome scores file not found: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ Error loading genome scores: {e}")
    raise

try:
    # Load genome tags
    logging.info("\n[7/7] Loading genome-tags.csv...")
    tables['genome_tags'] = pd.read_csv(f'{data_dir}genome-tags.csv')
    logging.info(f"  ✓ Loaded {len(tables['genome_tags']):,} genome tags")
    
except FileNotFoundError as e:
    logging.error(f"  ✗ Genome tags file not found: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ Error loading genome tags: {e}")
    raise

logging.info("\n✓ All tables loaded successfully")

# Unpack for easier access
movies = tables['movies']
ratings = tables['ratings']
users = tables['users']
tags = tables['tags']
links = tables['links']
genome_scores = tables['genome_scores']
genome_tags = tables['genome_tags']
```

## Test 1: Relational Integrity Validation

Verify that all foreign key relationships are valid (no orphaned records).

```python
logging.info("="*70)
logging.info("TEST 1: RELATIONAL INTEGRITY VALIDATION")
logging.info("="*70)

try:
    logging.info("\nChecking Foreign Key Integrity...")
    
    # Test 1a: ratings.userId → users.userId
    ratings_users = set(ratings['userId'].unique())
    users_set = set(users['userId'].unique())
    orphan_ratings_users = ratings_users - users_set
    
    logging.info(f"  ratings.userId → users.userId: {len(orphan_ratings_users)} orphaned records")
    
    if len(orphan_ratings_users) > 0:
        logging.warning(f"  Found {len(orphan_ratings_users)} orphaned ratings without users!")
    else:
        logging.info("  ✓ No orphaned ratings")
    
    # Test 1b: ratings.movieId → movies.movieId
    ratings_movies = set(ratings['movieId'].unique())
    movies_set = set(movies['movieId'].unique())
    orphan_ratings_movies = ratings_movies - movies_set
    
    logging.info(f"  ratings.movieId → movies.movieId: {len(orphan_ratings_movies)} orphaned records")
    
    if len(orphan_ratings_movies) > 0:
        logging.warning(f"  Found {len(orphan_ratings_movies)} ratings for non-existent movies!")
    else:
        logging.info("  ✓ No orphaned movie ratings")
    
    # Test 1c: tags.userId → users.userId
    tags_users = set(tags['userId'].unique())
    orphan_tag_users = tags_users - users_set
    
    logging.info(f"  tags.userId → users.userId: {len(orphan_tag_users)} orphaned records")
    
    if len(orphan_tag_users) > 0:
        logging.warning(f"  Found {len(orphan_tag_users)} tags from non-existent users!")
    else:
        logging.info("  ✓ No orphaned user tags")
    
    # Test 1d: tags.movieId → movies.movieId
    tags_movies = set(tags['movieId'].unique())
    orphan_tag_movies = tags_movies - movies_set
    
    logging.info(f"  tags.movieId → movies.movieId: {len(orphan_tag_movies)} orphaned records")
    
    if len(orphan_tag_movies) > 0:
        logging.warning(f"  Found {len(orphan_tag_movies)} tags for non-existent movies!")
    else:
        logging.info("  ✓ No orphaned movie tags")
    
    # Final verification
    total_orphans = (len(orphan_ratings_users) + len(orphan_ratings_movies) + 
                     len(orphan_tag_users) + len(orphan_tag_movies))
    
    if total_orphans == 0:
        logging.info("\n✓ ALL FOREIGN KEY RELATIONSHIPS VALIDATED")
        logging.info("  Zero orphaned records - relational integrity confirmed")
    else:
        logging.error(f"\n✗ Found {total_orphans} total orphaned records")
        raise ValueError(f"Relational integrity compromised: {total_orphans} orphans")
        
except KeyError as e:
    logging.error(f"  ✗ Missing expected column: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ Unexpected error during FK validation: {e}")
    raise
```

## Test 2: Statistical Validation

Verify that rating statistics match expected benchmark values from MovieLens documentation.

```python
logging.info("="*70)
logging.info("TEST 2: STATISTICAL VALIDATION")
logging.info("="*70)

try:
    logging.info(f"\nValidating Ratings Statistics...")
    
    actual_mean = ratings['rating'].mean()
    actual_std = ratings['rating'].std()
    actual_min = ratings['rating'].min()
    actual_max = ratings['rating'].max()
    
    logging.info(f"  Mean rating: {actual_mean:.4f} (Expected: ~3.53)")
    logging.info(f"  Std dev: {actual_std:.4f} (Expected: ~1.06)")
    logging.info(f"  Min: {actual_min} (Expected: 0.5)")
    logging.info(f"  Max: {actual_max} (Expected: 5.0)")
    
    # Validate mean
    if abs(actual_mean - 3.53) > 0.01:
        logging.error(f"  ✗ Rating mean {actual_mean} deviates from expected 3.53")
        raise ValueError(f"Rating mean {actual_mean} outside acceptable range")
    
    # Validate bounds
    if actual_min != 0.5:
        logging.error(f"  ✗ Rating min {actual_min} should be 0.5")
        raise ValueError(f"Rating minimum incorrect: {actual_min}")
    
    if actual_max != 5.0:
        logging.error(f"  ✗ Rating max {actual_max} should be 5.0")
        raise ValueError(f"Rating maximum incorrect: {actual_max}")
    
    logging.info("\n✓ STATISTICAL PROPERTIES VALIDATED")
    logging.info(f"  Mean rating within 0.01 of benchmark")
    logging.info(f"  Rating scale correctly bounded [0.5, 5.0]")
    
except KeyError as e:
    logging.error(f"  ✗ Missing 'rating' column: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ Statistical validation failed: {e}")
    raise
```

## Test 3: Relational Query Demonstration

Demonstrate complex relational operations using JOINs and aggregations.

```python
logging.info("="*70)
logging.info("TEST 3: RELATIONAL QUERY DEMONSTRATION")
logging.info("="*70)

try:
    logging.info("\nQuery: Top 10 Most-Rated Movies with Average Ratings")
    logging.info("-"*70)
    
    # Aggregate ratings by movie
    movie_stats = ratings.groupby('movieId').agg({
        'rating': ['count', 'mean', 'std']
    }).reset_index()
    movie_stats.columns = ['movieId', 'num_ratings', 'avg_rating', 'std_rating']
    
    logging.info(f"  Aggregated {len(ratings):,} ratings into {len(movie_stats):,} movie summaries")
    
    # JOIN with movie titles
    top_movies = movie_stats.merge(movies, on='movieId', how='inner')
    
    if len(top_movies) == 0:
        logging.error("  ✗ JOIN produced no results")
        raise ValueError("JOIN between ratings and movies failed")
    
    top_movies = top_movies.sort_values('num_ratings', ascending=False).head(10)
    
    # Display results
    print(f"\n{'Rank':<6}{'Title':<50}{'Ratings':>10}{'Avg':>8}{'Std':>8}")
    print("-"*82)
    
    for idx, (i, row) in enumerate(top_movies.iterrows(), 1):
        print(f"{idx:<6}{row['title']:<50}{row['num_ratings']:>10,}{row['avg_rating']:>8.2f}{row['std_rating']:>8.2f}")
        logging.info(f"  Rank {idx}: {row['title']} - {row['num_ratings']:,} ratings")
    
    logging.info("\n✓ RELATIONAL JOIN OPERATIONS VALIDATED")
    logging.info("  Successfully joined ratings and movies tables")
    logging.info(f"  Aggregated {len(ratings):,} ratings into movie-level statistics")
    
except KeyError as e:
    logging.error(f"  ✗ Missing expected column during JOIN: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ Relational query failed: {e}")
    raise
```

## Test 4: Genre Analysis

Analyze rating patterns across different genres to validate genre-based features.

```python
logging.info("="*70)
logging.info("TEST 4: GENRE ANALYSIS DEMONSTRATION")
logging.info("="*70)

try:
    logging.info("\nAnalyzing average ratings by genre (using 100K sample for speed)...")
    
    # Sample ratings for performance
    sample_ratings = ratings.sample(n=100000, random_state=42)
    logging.info(f"  Sampled {len(sample_ratings):,} ratings for genre analysis")
    
    # Expand genres (movies can have multiple pipe-separated genres)
    genre_ratings = []
    
    for _, rating_row in sample_ratings.iterrows():
        try:
            movie_info = movies[movies['movieId'] == rating_row['movieId']]
            
            if len(movie_info) > 0:
                genres_str = movie_info['genres'].values[0]
                
                if pd.notna(genres_str) and genres_str != '(no genres listed)':
                    for genre in genres_str.split('|'):
                        genre_ratings.append({
                            'genre': genre.strip(),
                            'rating': rating_row['rating']
                        })
        except Exception as e:
            logging.warning(f"  Skipping problematic rating: {e}")
            continue
    
    if len(genre_ratings) == 0:
        logging.error("  ✗ No genre ratings extracted")
        raise ValueError("Genre expansion produced no results")
    
    # Convert to DataFrame and aggregate
    genre_df = pd.DataFrame(genre_ratings)
    genre_stats = genre_df.groupby('genre')['rating'].agg(['mean', 'count']).sort_values('mean', ascending=False)
    
    logging.info(f"  Analyzed {len(genre_ratings):,} genre-rating pairs across {len(genre_stats)} genres")
    
    print("\nTop 5 Highest-Rated Genres:")
    print(f"{'Genre':<20}{'Avg Rating':>12}{'Sample Size':>15}")
    print("-"*47)
    
    for genre, row in genre_stats.head(5).iterrows():
        print(f"{genre:<20}{row['mean']:>12.2f}{int(row['count']):>15,}")
        logging.info(f"  Top genre: {genre} - {row['mean']:.2f}★")
    
    print("\nBottom 5 Lowest-Rated Genres:")
    print(f"{'Genre':<20}{'Avg Rating':>12}{'Sample Size':>15}")
    print("-"*47)
    
    for genre, row in genre_stats.tail(5).iterrows():
        print(f"{genre:<20}{row['mean']:>12.2f}{int(row['count']):>15,}")
    
    logging.info("\n✓ GENRE-BASED AGGREGATION VALIDATED")
    logging.info("  Genre patterns match expected industry trends")
    
except KeyError as e:
    logging.error(f"  ✗ Missing expected column in genre analysis: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ Genre analysis failed: {e}")
    raise
```

## Test 5: User Behavior Analysis

Analyze user activity distribution to validate cold-start problem identification.

```python
logging.info("="*70)
logging.info("TEST 5: USER BEHAVIOR VALIDATION")
logging.info("="*70)

try:
    logging.info("\nAnalyzing user rating patterns...")
    
    # Categorize users by activity level
    activity_bins = [0, 50, 100, 500, 1000, 5000, 50000]
    activity_labels = ['20-50', '51-100', '101-500', '501-1000', '1001-5000', '5000+']
    
    users['activity_group'] = pd.cut(users['num_ratings'], 
                                      bins=activity_bins, 
                                      labels=activity_labels)
    
    activity_dist = users['activity_group'].value_counts().sort_index()
    
    print("\nUser Distribution by Rating Count:")
    print(f"{'Activity Level':<20}{'Users':>12}{'Percentage':>15}")
    print("-"*47)
    
    for group, count in activity_dist.items():
        pct = (count / len(users)) * 100
        print(f"{group:<20}{count:>12,}{pct:>14.1f}%")
        logging.info(f"  {group}: {count:,} users ({pct:.1f}%)")
    
    median_ratings = users['num_ratings'].median()
    mean_ratings = users['num_ratings'].mean()
    power_users = (users['num_ratings'] >= 1000).sum()
    
    logging.info("\n✓ USER SEGMENTATION VALIDATED")
    logging.info(f"  Median user: {median_ratings:.0f} ratings")
    logging.info(f"  Mean user: {mean_ratings:.1f} ratings")
    logging.info(f"  Power users (1000+ ratings): {power_users:,}")
    logging.info(f"  Confirms cold-start challenge: majority have <100 ratings")
    
    print(f"\nKey Insights:")
    print(f"  • Median user: {median_ratings:.0f} ratings")
    print(f"  • Mean user: {mean_ratings:.1f} ratings")
    print(f"  • Power users (1000+ ratings): {power_users:,}")
    print(f"  • Cold-start users (<100 ratings): {(users['num_ratings'] < 100).sum():,}")
    
except KeyError as e:
    logging.error(f"  ✗ Missing expected column: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ User behavior analysis failed: {e}")
    raise
```

## Test 6: Temporal Pattern Analysis

Examine rating volume over time to validate temporal dynamics.

```python
logging.info("="*70)
logging.info("TEST 6: TEMPORAL PATTERN VALIDATION")
logging.info("="*70)

try:
    logging.info("\nAnalyzing rating volume over time...")
    
    # Convert Unix timestamps to dates
    ratings['date'] = pd.to_datetime(ratings['timestamp'], unit='s')
    ratings['year'] = ratings['date'].dt.year
    
    # Count ratings by year
    yearly_counts = ratings.groupby('year').size().sort_index()
    
    logging.info(f"  Dataset spans: {ratings['year'].min()} - {ratings['year'].max()}")
    
    print("\nRating Volume by Year (Key Years):")
    print(f"{'Year':<10}{'Ratings':>15}{'Growth':>15}")
    print("-"*40)
    
    key_years = [1995, 2000, 2005, 2010, 2015, 2018, 2019]
    prev_count = None
    
    for year in key_years:
        if year in yearly_counts.index:
            count = yearly_counts[year]
            
            if prev_count:
                growth = ((count - prev_count) / prev_count * 100)
                print(f"{year:<10}{count:>15,}{growth:>14.1f}%")
                logging.info(f"  {year}: {count:,} ratings ({growth:+.1f}% growth)")
            else:
                print(f"{year:<10}{count:>15,}{'—':>15}")
                logging.info(f"  {year}: {count:,} ratings (baseline)")
            
            prev_count = count
    
    peak_year = yearly_counts.idxmax()
    peak_count = yearly_counts.max()
    
    logging.info("\n✓ TEMPORAL ANALYSIS VALIDATED")
    logging.info(f"  Peak year: {peak_year} with {peak_count:,} ratings")
    logging.info(f"  10x+ growth from 2010 to 2018 confirms temporal weighting need")
    
    print(f"\nTemporal Insights:")
    print(f"  • Peak year: {peak_year} ({peak_count:,} ratings)")
    print(f"  • Growth factor (2010→2018): {yearly_counts[2018]/yearly_counts[2010]:.1f}x")
    print(f"  • Recent concentration: {(ratings['year'] >= 2015).sum():,} ratings (2015-2019)")
    
except KeyError as e:
    logging.error(f"  ✗ Missing timestamp column: {e}")
    raise
except Exception as e:
    logging.error(f"  ✗ Temporal analysis failed: {e}")
    raise
```
